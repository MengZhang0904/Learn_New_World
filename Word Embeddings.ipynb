# Shuffle

```python
shuffled = []
for i in range(1000):
    print('Shuffle time no. ' + str(i))
    for item in read_word_list:
        shuffle(item)
        item_copy = copy.deepcopy(item)
        shuffled.append(item_copy)
print(len(shuffled))

entire_shu1000 = 'entire_shuffle_1000.csv'
with open(entire_shu1000,'w') as output:
    writer = csv.writer(output,lineterminator='\n')
    writer.writerows(shuffled)
    
# train model, takes 24 mins.
model = Word2Vec(shuffled,size=100, window=5, min_count=1, workers=4)
# fit a 2d PCA model to the vectors
X = model[model.wv.vocab]
pca = PCA(n_components=2)
result = pca.fit_transform(X)
# create a scatter plot of the projection
plt.scatter(result[:, 0], result[:, 1])
words = list(model.wv.vocab)
for i, word in enumerate(words):
    plt.annotate(word, xy=(result[i, 0], result[i, 1]))
plt.show()

model.save('w2v_entire_s1000.bin')
```

```python
def myshuffle(ls,sft):
    shuffled = []
    for i in range(sft):
        print('Shuffle No.', i)
        for item in ls:
            shuffle(item)
            item_copy = copy.deepcopy(item)
            shuffled.append(item_copy)
    print('Shuffle is Done.' + 'Total length: ' + str(len(shuffled)))
    return shuffled
```

# Train Word Embedding for Each Age Group

```python
def spec_read2(m):
    file = str(m) + 'mo_shuffle_1000.csv'
    with open(file, 'r') as f:
      reader = csv.reader(f)
      ls = list(reader)
    return ls 
```

```python
def word_embedding(ls,n_com=2):
    # train model, takes 24 mins.
    model = Word2Vec(ls,size=100, window=5, min_count=1, workers=4)
    # fit a 2d PCA model to the vectors
    X = model[model.wv.vocab]
    pca = PCA(n_com)
    result = pca.fit_transform(X)
    # create a scatter plot of the projection
    plt.scatter(result[:, 0], result[:, 1])
    words = list(model.wv.vocab)
    for i, word in enumerate(words):
        plt.annotate(word, xy=(result[i, 0], result[i, 1]))
    plt.show()
    return model
```

```python
def spec_save2(model,m):
    file = 'w2v_model_' + str(m) + 'mo_s1000' + '.csv'
    export_to_file(model,file)
    
def spec_save3(model,m):
    file = 'w2v_model_' + str(m) + 'mo_s1000' + '.bin'
    model.save(file)
```

```python
for m in range(16,31):
    print('Now training on mo.: ' + str(m))
    ls = spec_read2(m)
    word_embedding(ls,n_com=2)
    spec_save2(model,m)
```
